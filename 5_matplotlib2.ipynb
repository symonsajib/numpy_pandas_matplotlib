```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.figure(figsize = (8,5))

plt.title("Gas Prices Over Time (In USD)")

plt.plot(gas.Year, gas.USA, 'b*-', Label = "America")
plt.plot(gas.Year, gas.Canada, 'r*-', Label = "Canada")
plt.plot(gas.Year, gas['South Korea'],'g*-', Label = "South Korea")
plt.plot(gas.Year, gas.Australia, 'y*-', label = 'Australia')

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US$')
plt.legend()

plt.show()
```

```python
fifa = pd.read_csv("fifa_data.csv")
fifa.head(5)
```

```python
# Histogram

bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins = bins, color = "#adc5de")

plt.xticks(bins)

plt.ylabel("Number of Plyers")
plt.xlabel("Skill level")

plt.title("Distribution of Plyaer Skills in FIFA 2018")

plt.show()
```

```python
# Pie Chart
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]

right = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]

labels = ["left", "right"]
colors = ["#abcdef", "#abdcef"]

plt.pie([left, right], labels = labels, colors = colors, autopct = '%.2f%%')

plt.title("Foot Preference of Fifa 2018")

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
fifa.Weight
```

```python
plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

Weights = [light,light_medium, medium, medium_heavy, heavy]

labels = ["Under 125", "125-150", "150-175", "175-200", "Over 200"]

explode = (.6,.05,.05,.05,.2)

plt.title("Weight Distribution")

plt.pie(Weights, labels = labels, autopct = "%0.2f%%", explode = explode, pctdistance = 0.6)
plt.show()
```

```python
plt.figure(figsize = (3,6))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']


labels = ['Barca', 'Madrid', 'Revs']

boxes = plt.boxplot([barcelona, madrid, revs], labels = labels)

for box in boxes['boxes']:
    box.set(color = 'blue', linewidth = 2)

plt.title("Soccar Teams Comparison")
plt.ylabel("Fifa Rating")

plt.show()
```
