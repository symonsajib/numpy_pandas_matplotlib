```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot([1,2,3],[2,4,6])
```

```python
x = [0,1,2,3, 4]
y = [0,2,4,6,8]

plt.plot(x,y, label = '2x', color = 'red', linestyle = '--',linewidth = 2, marker = '*', markersize = 10, markeredgecolor = 'blue')

plt.title("First Graph!", fontdict = {'fontname' : 'Comic Sans MS', 'fontsize': 20})

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.legend()
plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#color,marker,line
plt.plot(x,y, 'y*-', label = '2x')

plt.title("First Graph!", fontdict = {'fontname' : 'Comic Sans MS', 'fontsize': 20})

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.legend()
plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize the graph
plt.figure(figsize = (3,2), dpi = 100)

#color,marker,line
plt.plot(x,y, 'r*-', label = '2x')

## Line Number Two
x2 =np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'b', label = "x^2")
plt.plot(x2[5:], x2[5:]**2, 'b--' )

plt.title("First Graph!", fontdict = {'fontname' : 'Comic Sans MS', 'fontsize': 12})

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8])

plt.legend()

plt.savefig('mygraph.png', dpi = 100)

plt.show()
```

```python
# Bar Chart

labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)

patterns = ['/', '*', 'o']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```
